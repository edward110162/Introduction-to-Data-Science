```
import matplotlib.pyplot as plt
import pandas as pd

myData = pd.read_csv("https://drive.google.com/uc?export=download&id=1qsJJf3w9JIZJy1GDZfVE9yz3noHpSmac")
myData.shape

info_year = "資料年份: 97~107 其中一個"
info_search = "資料類別: Government tax, Government spending, Central subsidy"
info_graph = "圖表類型: bar(柱狀圖)"
#info_graph = "圖表類型: bar(柱狀圖), plot(折線圖)"

print(info_year)
input_year = int(input("▶ 請輸入年份: "))

if input_year <97 or  input_year >107:
  error_year = "未知的年份" + info_year
  print(error_year)
else:
  myData_year = myData[myData["year"] == input_year]
  input_year_str=str(input_year)

  print(info_search)
  input_search = str(input("▶ 請輸入查詢資料: "))
  if input_search != "Government tax" and input_search != "Government spending" and input_search != "Central subsidy":
    error_search = "未知的資料" + info_search
    print(error_search)
  else:
    myData_year = myData[myData["year"] == input_year]
    input_year_str=str(input_year)
  
    print(info_graph)
    input_graph = str(input("▶ 請輸入圖表類型: "))
    if input_graph == "bar":
      plt.bar(myData_year["city"], myData_year[input_search], width = 0.4, facecolor='#9999ff', edgecolor='white')
      plt.title("City " + input_search + ", " + input_year_str)
      plt.xlabel("City Name")
      plt.ylabel(input_search)
      plt.show()
    #elif input_graph == "plot":
    #  plt.plot(myData_year["city"], myData_year[input_search])
    #  plt.legend()
    #  plt.show()
    else:
      error_graph = "未知的圖表" + info_graph
      print(error_graph)
```
